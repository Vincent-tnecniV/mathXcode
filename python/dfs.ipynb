```python
import copy
import random
x = [[0]*8, [0]*8, [0]*8, [0]*8, [0]*8, [0]*8, [0]*8, [0]*8]
b = [[True]+[False]*7, [False]*8, [False]*8, [False]*8, [False]*8, [False]*8, [False]*8, [False]*8]
# pos- pos1 dir
stats_ = [0, 0, 0]
dirs = [(0, 1), (1, 0), (0, -1), (-1, 0)]
threshold = 107
prob = 0.000001
```

```python

def dfs(stats):
    # print(stats)
    # print(x)
    # print(b)
    # occupied
    if b[stats[0]][stats[1]]:
        # print
        # rt
        new_stats = [0, 0, stats[2]]
        if x[stats[0]][stats[1]] == 1:
            new_stats[2] += 1
        new_stats[0] = stats[0] + dirs[new_stats[2]%4][0]
        new_stats[1] = stats[1] + dirs[new_stats[2]%4][1]

        if 0 <= new_stats[0] < len(x[0]) and 0 <= new_stats[1] < len(x):
            dfs(new_stats)
        else:
            if stats[2] > threshold:
                print("\n\n")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])
            elif random.random() < prob/10:
                print("\n\nnot optimistic")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])
            else:
                if random.random() < prob:
                    print('.', end='')

    # not occupied
    else:
        b[stats[0]][stats[1]] = True
        # cheese
        x[stats[0]][stats[1]] = 1
        new_stats = [0, 0, stats[2]+1]
        new_stats[0] = stats[0] + dirs[new_stats[2]%4][0]
        new_stats[1] = stats[1] + dirs[new_stats[2]%4][1]
        if 0 <= new_stats[0] < len(x[0]) and 0 <= new_stats[1] < len(x):
            dfs(new_stats)
        else:

            if stats[2] > threshold:
                print("\n\n")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])
            elif random.random() < prob/10:
                print("\n\nnot optimistic")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])

            else:
                if random.random() < prob:
                    print('.', end='')
        # no cheese
        x[stats[0]][stats[1]] = 0

        new_stats = [0, 0, stats[2]]
        new_stats[0] = stats[0] + dirs[new_stats[2]%4][0]
        new_stats[1] = stats[1] + dirs[new_stats[2]%4][1]
        if 0 <= new_stats[0] < len(x[0]) and 0 <= new_stats[1] < len(x):
            dfs(new_stats)
        else:

            if stats[2] > threshold:
                print("\n\n")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])
            elif random.random() < prob/10:
                print("\n\nnot optimistic")
                print(stats)
                # print(b)
                print(x[0])
                print(x[1])
                print(x[2])
                print(x[3])
                print(x[4])
                print(x[5])
                print(x[6])
                print(x[7])
            else:
                if random.random() < prob:
                    print('.', end='')


        b[stats[0]][stats[1]] = False
```
